```python
#required imports
import pandas as pd
import warnings
import io
import numpy as np
warnings.filterwarnings('ignore')

from google.colab import drive
from google.colab import files


uploaded = files.upload()

```

```python
# Q1. Read the housingdata.csv file into pandas DataFrame and display first six rows of the DataFrame
hdata=pd.read_csv(io.BytesIO(uploaded['housingdata.csv']))
hdata.head(6)
```

```python
# Q2. Display the column names and row index in separate cells.
column_names=list(hdata.columns)
column_names.insert(0,'')
hdata2=pd.DataFrame(columns=column_names)
hdata2=hdata2.append(hdata)
hdata2.reset_index(inplace=True)
hdata2.rename(columns={'index':'Row Index'} , inplace = True )
hdata2
```

```python
# Q3. a) How many numbers of “STATE” exist in the dataset. Hint: use shape
print(hdata['STATE'].shape)
#Q3. b) How many unique states exist in the dataset
print(hdata['STATE'].unique())
```

```python
#Q4. Retrieve the list of all NaN/Null/Empty cells in the form of Boolean list
Boolean_Hdata=hdata.isna()|hdata.isnull()|hdata.eq('')
print(Boolean_Hdata)
```

```python
#Q5. Drop all rows with N/A,NA,na values in Num_Bedrooms
hdata=hdata[~hdata['NUM_BEDROOMS'].isin(['na','NaN','N/A'])]
print(hdata)
```

```python
#Q6. Replace the NaN and String entries in Num_Bath with previous entry
hdata['NUM_BATH'] = pd.to_numeric(hdata['NUM_BATH'],errors='coerce')
hdata['NUM_BATH'] = hdata['NUM_BATH'].fillna(method='ffill')
print(hdata)
```

```python
#Q7. Replace the empty values in “SQ_FT” with the mean of the all entries.
hdata['SQ_FT']=hdata['SQ_FT'].replace('--',np.nan)
hdata['SQ_FT']=hdata['SQ_FT'].astype(float)
mean_sqft=hdata['SQ_FT'].mean()
hdata['SQ_FT'].fillna(mean_sqft,inplace=True)
print(hdata)
```

```python
#Q8. What is the sum of SQ_FT of all entries in the dataset.
hdata['SQ_FT'].sum()
```

```python
#Q9. Add a new column “NUM_STORE” and with all the values as 1 for all rows
hdata['NUM_STORE']=[1,1,1,1,1,1,1,1]
hdata
```

```python
#Q10. Create a DataFrame containing all entries of TN state only
hdata_tn=hdata[hdata['STATE']=='TN']
hdata_tn=hdata_tn.reset_index()

hdata_tn.drop('index',axis=1,inplace=True)
hdata_tn
```

```python
#Q14. Create a DataFrame having SQ_FT area greater than 1000 and display the DataFrame
hdata1=hdata[hdata['SQ_FT']>1000.0]
hdata1=hdata1.reset_index()
hdata1.drop('index',axis=1,inplace=True)
hdata1
```

```python
#Q15. Create a DataFrame having only first Three columns and First Three rows
hdata.iloc[0:3,0:3]
```

```python
#Q16. Display the state having highest average price per square foot of housing area.
state_prices=hdata.groupby('STATE')['PRICE','SQ_FT'].mean()
state_prices['pricepersqft']=state_prices['PRICE']/state_prices['SQ_FT']
state_prices=state_prices.sort_values('pricepersqft',ascending=True)
print('The state having highest average price per square foot of housing area is:',state_prices.index[0])
```
